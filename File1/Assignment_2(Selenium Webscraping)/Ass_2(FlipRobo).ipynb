Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

#### Importing all the required Liberaries

```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
```

#### WebDriver is an open source tool for automated testing of webapps across many browsers. 

```python
driver = webdriver.Chrome(r'chromedriver.exe') #WebDriver is an open source tool for automated testing of webapps across many browsers.
driver.get('https://www.naukri.com/') #The get command launches a new browser and opens the given URL in your Webdriver.
designation = driver.find_element(By.CLASS_NAME, 'suggestor-input') # find element by class 
designation.send_keys('Data Analyst') #write the designation as Data Analyst in designation column
location = driver.find_element(By.XPATH, '/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input') #write the location as Bangalore in location column 
location.send_keys('Bangalore')  #write the location as Bangalore in location column

button = driver.find_element(By.CLASS_NAME, 'qsbSubmit')
button.click()   #click the search button
```

##### scrape the job-title, job-location, company_name, experience_required.

```python
job = []                              #Making Differenet lists 
location = []
company = []
experience = []

job_title = driver.find_elements(By.XPATH, '//a[@class="title fw500 ellipsis"]') #find element with using relative XPATH       
for i in job_title[:10]:      #for top 10 results
    job_title = i.text        
    job.append(job_title)

job_location = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft fs12 lh16 locWdth"]')    #find element with using relative XPATH 
for i in job_location[:10]:
    job_location = i.text
    location.append(job_location)
    
        

company_name =  driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')    #find element with using relative XPATH 
for i in company_name[:10]:
    company_name = i.text
    company.append(company_name)  
    
experience_required = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft fs12 lh16 expwdth"]')  #find element with using relative XPATH 
for i in experience_required[:10]:
    experience_required = i.text
    experience.append(experience_required)
    
```

```python
Data_Analyst = pd.DataFrame({'Job_Title':job, 'Job_Location':location, 'Company_Name':company, 'Experience_Required':experience})   #Making DataFrame and put it into a variable Data_Analyst
```

Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price

To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url : https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data as usual.

4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100 sunglasses.

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.flipkart.com/')
sunglasses = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
sunglasses.send_keys('Sunglass')
enter = driver.find_element(By.CLASS_NAME, 'L0Z3Pu')
enter.click()
```

```python
element_list = []
  
for page in range(1, 4):
    brand_name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')  #find element with using relative XPATH 
    product = driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]') #find element with using relative XPATH 
    glass_name = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]' )#find element with using relative XPATH 
    
    
    for i in range(len(brand_name)):
         element_list.append([brand_name[i].text, product[i].text, glass_name[i].text])
            
#print(element_list)

driver.close()  #driver.close() closes only the current window on which Selenium is running automated tests
```

```python
pd.DataFrame(element_list, columns = ['Sunglass_brand', 'Description', 'Price'])[:100]  
```

As shown in the above page you have to scrape the tick marked attributes. These are:

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')


element_list = []

for i in range(1,11):
    rating = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')[:100]  #find element with using relative XPATH 
    review_summ= driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')[:100]   #find element with using relative XPATH 
    full_review = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')[:100]   #find element with using relative XPATH 
    
    for i in range(len(rating)):
        element_list.append([rating[i].text, review_summ[i].text, full_review[i].text])
#print(element_list)        
driver.close()  

```

### Q6: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 4 attributes of each sneaker:
1. Brand
2. Product Description
3. Price



```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.flipkart.com/')
search = driver.find_element(By.CLASS_NAME, '_3704LK')
search.send_keys('sneakers')
button = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
button.click()
```

```python
element_list = []

for i in range(1,4):
    brand = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')[:100]
    price = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')[:100]
    desc = driver.find_elements(By.XPATH, '//a[contains(@class,"IRpwTa") or contains(@class,"IRpwTa _2-ICcC")]')[:100]      # 'or' is used for either the element present among these 2 classes     
    
    for i in range(len(brand)):
        element_list.append([brand[i].text, price[i].text, desc[i].text])
#print(element_list)        
driver.close()  
```

```python
sneakers = pd.DataFrame(element_list, columns = ['Brand_Name', 'Price', 'Description'])[:100]
sneakers
```
