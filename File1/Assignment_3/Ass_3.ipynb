4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.flipkart.com/search?q=samsung%20galaxy%20z&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
```

```python
lists = []
for i in range(0,1):
    name = driver.find_elements(By.XPATH, '//div[@class="_4rR01T"]')
    storage = driver.find_elements(By.XPATH, '//ul[@class="_1xgFaf"]')    
    #camera = driver.find_element(By.XPATH, '')
    #display_size = driver.find_element(By.XPATH, '')
    #battery_capacity = driver.find_element(By.XPATH, '')
    price = driver.find_elements(By.XPATH, '//div[@class="_30jeq3 _1_WHN1"]')
    
    for i in range(len(name)):
        lists.append([name[i].text, price[i].text])
```

```python
data_1 = pd.DataFrame(lists, columns = ['Product_Name', 'Price'])
```

```python
conf = pd.DataFrame([i.text for i in storage], columns = ['config'])
conf = conf['config'].str.split('\n', expand=True)
conf.rename(columns = {0:'Storage', 1:'Display', 2:'Camera', 3:'Battery', 4:'Processor', 5:'Warranty'}, inplace = True)
conf = conf.iloc[:,0:4]
```

```python
phone_data = pd.concat([data_1,conf], axis = 1)
#phone_data.to_csv('Phone_Data')
```

8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://www.forbes.com/billionaires/')
```

```python
lists = []

for page in range(1,4):
    rank = driver.find_elements(By.XPATH, '//div[@class="rank"]')
    name = driver.find_elements(By.XPATH, '//div[@class="personName"]')
    worth = driver.find_elements(By.XPATH, '//div[@class="netWorth"]')
    age = driver.find_elements(By.XPATH, '//div[@class="age"]')
    citizen = driver.find_elements(By.XPATH, '//div[@class="countryOfCitizenship"]')
    source = driver.find_elements(By.XPATH, '//div[@class="source"]')
    industry = driver.find_elements(By.XPATH, '//div[@class="category"]') 
    
    for i in range(len(rank)):
        lists.append([rank[i].text, name[i].text, worth[i].text, age[i].text, citizen[i].text, source[i].text, industry[i].text])
        
```

```python
for i in range(len(name)):
    lists.append([name[i].text, distance[i].text, rating[i].text, total_review[i].text, facilities[i].text,overall_reviews[i].text, price[i].text])
data_1 = pd.DataFrame(lists, columns = ['hostel_name', 'distance', 'ratings', ' total_review', 'facilities','overall_review', 'price']) 
Desc = []
while True:
    hostel_url=urls =[i.get_attribute('href') for i in driver.find_elements(By.XPATH,"//h2[@class='title title-6']//a")]
    try:
        next_button = driver.find_elements(By.XPATH,'//div[@class="pagination-item pagination-next"]').click()
    except:
        break
for url in hostel_url:
    driver.get(url)
    time.sleep(2) 
    try:        
        desc = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[2]/div/div/div[2]')
        Desc.append(desc.text)        
    except NoSuchElementException:
        Desc.append('_') 
```
